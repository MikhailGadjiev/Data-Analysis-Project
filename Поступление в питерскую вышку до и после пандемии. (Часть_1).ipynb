# Cluster Analysis

```python
from sklearn.metrics import calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.stats import norm
from scipy import stats
df_clstr=df.dropna(subset=['Russian_Lan_EGE_2020','Foreign_Lan_EGE_2020', 
                           'Extra_Awards_2020','Social_Sc_EGE_2019','Mathematics_EGE_2020'])
```

```python
import warnings
warnings.filterwarnings("ignore")
df_clstr['z_Russian_Lan_EGE_2020'] = (df_clstr['Russian_Lan_EGE_2020'] - 
                                      df_clstr['Russian_Lan_EGE_2020'].mean())/df_clstr['Russian_Lan_EGE_2020'].std()
df_clstr['z_Foreign_Lan_EGE_2020'] = (df_clstr['Foreign_Lan_EGE_2020'] - 
                                      df_clstr['Foreign_Lan_EGE_2020'].mean())/df_clstr['Foreign_Lan_EGE_2020'].std()
df_clstr['z_Extra_Awards_2020'] = (df_clstr['Extra_Awards_2020'] - 
                                   df_clstr['Extra_Awards_2020'].mean())/df_clstr['Extra_Awards_2020'].std()
df_clstr['z_Social_Sc_EGE_2020'] = (df_clstr['Social_Sc_EGE_2020'] - 
                                    df_clstr['Social_Sc_EGE_2020'].mean())/df_clstr['Social_Sc_EGE_2020'].std()
df_clstr['z_Mathematics_EGE_2020'] = (df_clstr['Mathematics_EGE_2020'] - 
                                      df_clstr['Mathematics_EGE_2020'].mean())/df_clstr['Mathematics_EGE_2020'].std()

#I want to standardize the values
```

```python
plt.figure(figsize=(15, 6))
dn = dendrogram(linkage(df_clstr[['z_Russian_Lan_EGE_2020', 'z_Social_Sc_EGE_2020', 'z_Mathematics_EGE_2020', 'z_Extra_Awards_2020','z_Foreign_Lan_EGE_2020']], 'ward'))
#Ward's Method. According to this approach, we have 3 clusters.
```

```python
plt.figure(figsize=(15, 6))
dn = dendrogram(linkage(df_clstr[['z_Russian_Lan_EGE_2020', 'z_Social_Sc_EGE_2020', 'z_Mathematics_EGE_2020', 'z_Extra_Awards_2020','z_Foreign_Lan_EGE_2020']], 'single'))
#Single linkage Method. I can't detect a particular number in this method.
```

```python
plt.figure(figsize=(15, 6))
dn = dendrogram(linkage(df_clstr[['z_Russian_Lan_EGE_2020', 'z_Social_Sc_EGE_2020', 'z_Mathematics_EGE_2020', 'z_Extra_Awards_2020','z_Foreign_Lan_EGE_2020']], 'centroid'))
#Centroid Method. I can't detect a particular number in this method.
```
